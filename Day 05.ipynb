```python
### problem 01
length = int(input('Enter the length of a rectangle: '))
width = int(input('Enter the width of a rectangle: '))
if length == width:
    print("It's a square")
else:
    print ("It's not a square")
```

```python
#problem 02
num1 = int(input('Enter first number:'))
num2 = int(input('Enter second number:'))
num3 = int(input('Enter third number:'))
if (num1 > num2) and (num1 > num3):
    largest = num1
elif (num2 > num1) and (num2 > num3):
    largest = num2
else:
    largest = num3
print(largest,'is greatest among of them')
```

```python
#problem 03
attend = int(input('Enter your attendance day:'))
total_class = int(input('Enter total days of class:'))
attendance = (attend / total_class) * 100
if attendance >= 75:
    print('You can sit in the exam')
else:
    print('You can not sit in the exam')
```

```python
#problem 04
marks = int(input('Enter your marks: '))
if marks < 25:
    print('F Grade')
elif marks >= 25 and marks <= 44:
    print('E Grade')
elif marks >= 45 and marks <= 49:
    print('D Grade')
elif marks >= 50 and marks <= 59:
    print('C Grade')
elif marks >= 60 and marks <= 79:
    print('B Grade')
elif marks >= 80 and marks <= 89:
    print('A Grade')
else:
    print('A+ Grade')
```

```python
num = float(input('Enter the number: '))
if num % 2 == 0:
    if num % 5 == 0:
        print('Divisible by 5 and 2')
    else:
        print('Divisible by 2 only')
else:
    print('Divisible not by 2 or 5')
```
