```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
[x for x in range(10)]
```

```python
for x in range(10):
    print(x)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
odds = [x for x in range(10) if x % 2 != 0]
print(odds)
```

```python
data = 'I love data science'
[x for x in data]
```

```python
data = 'I love data science'
data = data.split()
[x for x in data]
```

```python
[x for x in data if x != 'I']
```
